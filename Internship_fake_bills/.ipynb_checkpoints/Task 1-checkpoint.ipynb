```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('C:/Users/Lenovo/OneDrive/Desktop/fake_bills.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
quant=0
quali=0
nom=0
cat=0
for column in data.columns:
    if data[column].dtype in ['int64', 'float64']:
        quant+= 1
    else:  # Non-numeric data types
        quali += 1
        unique_values = len(data[column].unique())
        if unique_values>2:
            nom+= 1
        else:
            cat+= 1
```

```python
print(f"Quantitative:{quant}")
print(f"Quanlitative:{quali}")
print(f"Nominal:{nom}")
print(f"Categorical:{cat}")
```

```python
print('Display the descriptive statics')
all_stats = data.describe(include='all')
print(all_stats)
```
